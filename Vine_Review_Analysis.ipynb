```python
import pandas as pd
import os
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df
```

```python
filtered_df = vine_df[vine_df['total_votes']>20]
filtered_df
```

```python
helpful_df = filtered_df[filtered_df['helpful_votes']/filtered_df['total_votes'] >= 0.5]
helpful_df
```

```python
vine_y_df = helpful_df[helpful_df['vine']=='Y']
vine_n_df = helpful_df[helpful_df['vine']=='N']
```
